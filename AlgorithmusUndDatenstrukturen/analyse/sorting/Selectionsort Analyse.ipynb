```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Set figure width to 12 and height to 9
fig_size[0] = 15
fig_size[1] = 12
plt.rcParams["figure.figsize"] = fig_size
```

## Loading Data

```python
raw_data = pd.read_csv('data/SortAnalyse.csv', header=None)

raw_data.columns = ['list_type', 'sort', 'permutation', 'size', 'comparison', 'time', 'pivot']

raw_data
```

### Grouping and hierarchical indexing 

```python
data = raw_data.groupby(['list_type', 'sort', 'permutation', 'size']).mean()
data
```

## Analyse FastInsertionSort

```python
is_random_y = data.loc['FastElements','FastInsertionSort','RandomPermutation']['time']
is_random_x = is_random_y.index.values
plt.plot(is_random_x, is_random_y, label='random')

is_reverse_y = data.loc['FastElements','FastInsertionSort','ReversePermutation']['time']
is_reverse_x = is_reverse_y.index.values
plt.plot(is_reverse_x, is_reverse_y, label='reverse')

is_almost_y = data.loc['FastElements','FastInsertionSort','AlmostSortedPermutation']['time']
is_almost_x = is_almost_y.index.values
plt.plot(is_almost_x, is_almost_y, label='almost sorted')

plt.legend()
plt.title('FastInsertionSort Analyse')
plt.xlabel('Größe der Liste')
plt.ylabel('Laufzeit in Sekunden')

plt.show()
```

## Analyse SelectionSort

```python
ss_random_y = data.loc['FastElements', 'SelectionSort', 'RandomPermutation']['time']
ss_random_x = ss_random_y.index.values
plt.plot(ss_random_x, ss_random_y, label='random')

ss_reverse_y = data.loc['FastElements', 'SelectionSort', 'ReversePermutation']['time']
ss_reverse_x = ss_reverse_y.index.values
plt.plot(ss_reverse_x, ss_reverse_y, label='reverse')

ss_almost_y = data.loc['FastElements', 'SelectionSort', 'AlmostSortedPermutation']['time']
ss_almost_x = ss_almost_y.index.values
plt.plot(ss_almost_x, ss_almost_y, label='almost sorted')

plt.legend()
plt.title('FastInsertionSort Analyse')
plt.xlabel('Größe der Liste')
plt.ylabel('Laufzeit in Sekunden')

plt.show()
```

## FastInsertionSort vs SelectionSort

```python
plt.plot(is_random_x, is_random_y, label='InsertionSort random')
plt.plot(is_reverse_x, is_reverse_y, label='InsertionSort reverse')
plt.plot(is_almost_x, is_almost_y, label='InsertionSort almost sorted')
plt.plot(ss_random_x, ss_random_y, label='SelectionSort')

plt.legend()
plt.title('FastInsertionSort vs SelectionSort')
plt.xlabel('Größe der Liste')
plt.ylabel('Laufzeit in Sekunden')

plt.show()
```
